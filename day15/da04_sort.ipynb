#### 퀵정렬
- 기준을 하나 뽑은 후 기준보다 작은 그룹과 큰 그룹을 나누어 다시 각 그룹으로 정렬하는 방법

```python
def sortQuick(ary):
    global count
    n = len(ary)
    if n <= 1 :     #정렬해야하는 리스트 길이가 1이하
        return ary
    
    pivot = ary[n//2]   #중앙값을 기준으로 함
    leftAry , rightAry= [] ,[]

    for num in ary :
        if num < pivot :
            leftAry.append(num)
            count +=1
        elif num == pivot : # 중앙값은 기준이기에 왼쪽, 오른쪽 배열에 넣지 않으므로 / recurssionError 방지위해 필요
            continue
        else :
            rightAry.append(num)
            count +=1

    return sortQuick(leftAry) +[pivot] + sortQuick(rightAry)    #재귀호출


dataAry =[188, 162, 198, 120, 50, 150, 177, 105]
count = 0
# 메인코드
print(f'정렬 전->{dataAry}')
dataAry = sortQuick(dataAry)
print(f'처리횟수->{count}')
print(f'정렬 후->{dataAry}')

```

- 중복값을 고려한 퀵 정렬

```python
def sortQuick(ary):
    n= len(ary)

    if n <= 1 : return ary

    pivot = ary[n//2]
    leftAry , rightAry , midAry = [] ,[], []

    for i in ary :
        if i >pivot :
            rightAry.append(i)
        elif i== pivot :
            midAry.append(i)
        else :
            leftAry.append(i)

    return sortQuick(leftAry) + midAry + sortQuick(rightAry)

dataAry = [188, 162, 198, 120, 50, 150, 177, 105 , 120,150]

print(f'정렬 전->{dataAry}')
dataAry = sortQuick(dataAry)
print(f'정렬 후->{dataAry}')
```

- 퀵 정렬 일반적 구현

```python
def sortQuickN(ary, start, end) :
    if end <= start :       # 배열이 홀수이면 end== start, 배열이 짝수이면 start > end
        return 
    
    low = start
    high = end
    pivot =ary [(low+high)//2 ]

    while low <= high :
        while ary[low] < pivot :
            low += 1
        while ary[high] >pivot :
            high -=1
        if low <= high :
            ary [low] , ary[high] = ary[high] , ary[low]
            low , high = low+1 , high-1
            print(f'정렬 중->{ary}')
    mid = low

    sortQuickN(ary, start, mid -1)
    sortQuickN(ary, mid, end)

# 전역변수

dataAry =[188, 150,168,162,105,120,177,50]

print(f'정렬 전->{dataAry}')
sortQuickN(dataAry, 0, len(dataAry)-1)
print(f'정렬 후->{dataAry}')

    
```
